## Visualize the Heart disease dataset by plotting the following graphs using R/Python.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv("Cleavland.csv", sep=",", header=None)
```

```python
df.head()
```

```python
col_names = {0:'age', 1:'sex', 2:'cp', 3:'trestbps', 4:'chol', 5:'fbs', 6:'restecg', 7:'thalach', 
             8:'exang', 9:'oldpeak', 10:'slope', 11:'ca', 12:'thal', 13:'num'}
df = df.rename(columns = col_names)
```

```python
df = df.dropna()
```

```python
df.head(5)
```

```python
for col in df.columns:
    df[col] = [x if x!='?' else np.nan for x in df[col]] 
df = df.dropna()
```

### a. Histograms

```python
sns.set()
fig, ax = plt.subplots(figsize=(15,6))
ax.hist(df.num)
plt.xlabel('Cancer chance')
plt.ylabel('No.of people')
plt.title('Cancer chance on people surveyed')
```

```python
ax = sns.distplot(df.num)
```

### b. Dot Plots

```python
df_ = df[:][:50]
plt.plot(df_.age, df_.num, 'ro')
```

```python
sns.pointplot(x="num", y="age", hue="sex", data=df)
```

### c. Bar Plots

```python
fig, ax = plt.subplots(figsize=(15, 6))
plt.bar(df.age, df.num)
plt.xlabel('Age')
plt.ylabel('Chance of Heart Disease')
plt.title('AGE VS HEART DISEASE')
plt.show()
```

```python
sns.countplot(data=df, x="num", hue="sex")
```

### d. Line Charts

```python
df_ = df[:][:5]
plt.plot(df.chol)
plt.show()
```

```python
sns.pointplot(x="num", y="age", hue="sex", data=df)
```

### b. Pie Charts

```python
sns.set()
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99', '#facb29']
plt.pie(df.num.value_counts(), labels=['0', '1', '2', '3', '4'], autopct='%1.1f%%', shadow=True, colors=colors, startangle=90)
plt.axis('equal')
plt.tight_layout()
```

```python
fig, ax = plt.subplots(figsize=(10,10))
df_ = df[['trestbps', 'chol']]
sns.boxplot(data=df_)
```

### d. Scatter Plots

```python
plt.scatter(df.age, df.num, marker="*", c="red", s=80, alpha=0.5)
```

```python
sns.lmplot(x="age", y="num", hue="sex", data=df, fit_reg=False)
```
