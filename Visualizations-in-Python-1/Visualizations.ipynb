## Visualizing with python

#### Histograms

```python
sns.set()
plt.hist(df['num'], color='blue')
plt.xlabel('Chance of heart disease')
plt.ylabel('No. of people')
plt.show()
```

#### Dot plots

```python
plt.plot(df['num'], 'ro')
plt.ylabel('Chance of heart disease')
plt.xlabel('No. of people')
plt.show()
```

#### Bar plots

```python
plt.bar(df['age'], df['num'])
plt.ylabel('CHANCE OF HEART DISEASE')
plt.xlabel('AGE')
plt.title('AGE AND HEART DISEASE')
plt.show()
```

#### Simple Plot

```python
plt.plot(df['age'], df['num'], 'r*')
plt.ylabel('Chance of heart disease')
plt.xlabel('AGE')
plt.show()
```

#### Line chart 

```python
plt.plot(df.chol)
plt.ylabel('CHOLESTROL LEVEL')
plt.xlabel('No. of people')
plt.show()
```

#### Pie chart

```python
sns.set()
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99', '#facb29']
plt.pie(df.num.value_counts(), labels=['0', '1', '2', '3', '4'], autopct='%1.1f%%', shadow=True, colors=colors, startangle=90)
plt.title('% of people having heart disease chances')
plt.axis('equal')
plt.tight_layout()
plt.show()
```

#### Scatter plot

```python
df_ = df[:][:50]
plt.scatter(df_.age, df_.num, label="agevHd", marker="*", s=80, color='red')
plt.ylabel('Chance of heart disease')
plt.xlabel('AGE')
plt.title("AGE VS HEART DISEASE")
plt.show()
```

#### Box plot

```python
plt.boxplot( df['thalach'])
df_ = df[:][:50]
plt.scatter(df_.age, df_.num, label="agevHd", marker="*", s=80, color='red')

plt.show()
```
